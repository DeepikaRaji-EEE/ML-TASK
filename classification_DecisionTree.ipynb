```python
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Actual and Predicted labels
actual_labels = [
    "Pass", "Pass", "Pass", "Pass",
    "Fail", "Fail", "Fail", "Fail", "Fail", "Fail"
]

predicted_labels = [
    "Pass", "Fail", "Pass", "Pass",
    "Fail", "Pass", "Fail", "Fail", "Fail", "Fail"
]

# Create Label Encoder
le = LabelEncoder()

# Fit encoder on all labels
le.fit(actual_labels + predicted_labels)

# Encode labels
actual = le.transform(actual_labels)
predicted = le.transform(predicted_labels)

print("Encoded Actual:", actual)
print("Encoded Predicted:", predicted)

# Confusion Matrix
cm = confusion_matrix(actual, predicted)
print("\nConfusion Matrix:")
print(cm)

# Evaluation Metrics
accuracy = accuracy_score(actual, predicted)
precision = precision_score(actual, predicted)
recall = recall_score(actual, predicted)
f1 = f1_score(actual, predicted)

print("\nAccuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
