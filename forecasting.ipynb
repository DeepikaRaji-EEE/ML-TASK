```python
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
# Stock price data
data = [200, 202, 205, 207, 210]
# Create ARIMA model
model = ARIMA(data, order=(2, 2, 2))
model_fit = model.fit()
# Forecast next day price
forecast = model_fit.forecast(steps=5)
print("Forecasted stock price for next day:", forecast[0])
print("Forecasted stock price for next day:", forecast[1])
print("Forecasted stock price for next day:", forecast[2])
```
