```python
import pandas as pd

data = {
    "Age": [22, None, 25, None, 35],
    "City": ["Delhi", None, "Mumbai", None, "Chennai"],
    "porfit": [500, 600, None, None, 750]
}

df = pd.DataFrame(data)
print(df)
```

```python
print(df.fillna(0))
```

```python
print(df["Age"].fillna(df["Age"].mean()))
```

```python
print(df["Age"].fillna(df["Age"].median()))
```

```python
print(df["City"].fillna(df["City"].mode()[0]))
```

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='bfill')
```

```python
print(df.fillna(0, inplace=True))
```
