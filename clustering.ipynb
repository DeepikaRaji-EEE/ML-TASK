```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

cups = np.array([[10], [0.15], [3.0], [28.0], [19.0], [10.0]])

model = KMeans(n_clusters=2, random_state=42)
model.fit(cups)

labels = model.labels_
centroids = model.cluster_centers_

plt.scatter(cups, labels)

plt.scatter(centroids, [0, 1], marker='X', s=500)

plt.xlabel("Cups of Tea per Day")
plt.ylabel("Cluster")
plt.title("K-Means Clustering ")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

data = np.array([[2.5], [8], [1.0], [9]])

Z = linkage(data, method='single')

plt.figure()
dendrogram(Z, labels=[2, 8, 10, 9])
plt.title("Hierarchical Clustering Dendrogram (Agglomerative)")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram

# Data: daily travel distance (km)
data = np.array([[2], [4], [6], [10]])

Z = linkage(data, method='single')

print("Linkage Matrix:")
print(Z)

plt.figure()
dendrogram(Z, labels=[2, 4, 6, 10])
plt.title("Divisive Hierarchical Clustering ")
plt.xlabel("Distance (km)")
plt.ylabel("Distance Between Clusters")
plt.show()
```
