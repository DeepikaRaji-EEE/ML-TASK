```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

# Sample Data
Experience = ["Senior", "Junior", "Senior", "Junior", "Senior", "Junior"]
Skill = ["Python", "Java", "Python", "Java", "Python", "Python"]
Hired = ["Yes", "No", "Yes", "No", "Yes", "No"]

# Create Label Encoders
exp_le = LabelEncoder()
skill_le = LabelEncoder()
hired_le = LabelEncoder()

# Encode data
Experience_enc = exp_le.fit_transform(Experience)
Skill_enc = skill_le.fit_transform(Skill)
Hired_enc = hired_le.fit_transform(Hired)

# Features and Target
X = list(zip(Experience_enc, Skill_enc))
y = Hired_enc

# Train Decision Tree
model = DecisionTreeClassifier(criterion="entropy")
model.fit(X, y)

# New Candidate Data
new_experience = "Senior"
new_skill = "Java"

# Encode new data
new_exp_enc = exp_le.transform([new_experience])
new_skill_enc = skill_le.transform([new_skill])

new_data = list(zip(new_exp_enc, new_skill_enc))

# Predict
prediction_enc = model.predict(new_data)

# Decode result
prediction = hired_le.inverse_transform(prediction_enc)

print("Candidate Hired?", prediction[0])
```
