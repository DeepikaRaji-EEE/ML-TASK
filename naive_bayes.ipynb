```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import LabelEncoder

# Different sample email dataset

emails = [
    "claim your free vacation now",
    "limited time offer just for you",
    "meeting scheduled for tomorrow",
    "can we discuss the project update",
    "earn cash quickly from home",
    "your invoice has been attached",
    "exclusive discount available today",
    "please review the presentation slides"
]

labels = [
    "Spam",
    "Spam",
    "Not Spam",
    "Not Spam",
    "Spam",
    "Not Spam",
    "Spam",
    "Not Spam"
]

# Encode output labels
le = LabelEncoder()
y = le.fit_transform(labels)

# Encode input text
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(emails)

# Train model
model = MultinomialNB()
model.fit(X, y)

# Test
test_email = ["claim limited"]
test_vec = vectorizer.transform(test_email)

pred = model.predict(test_vec)

# Decode result
print("Prediction:", le.inverse_transform(pred)[0])
```

```python
# Import libraries
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder

X = np.array([[55], [58], [62], [85], [92]])

y_text = ["Average", "Average", "Good", "Excellent", "Excellent"]

# Encode labels (Fail=0, Pass=1)
le = LabelEncoder()
y = le.fit_transform(y_text)

model = GaussianNB()


model.fit(X, y)

new_marks = [[65]]
prediction = model.predict(new_marks)

# Decode result
print("Predicted Result for 65 marks:", le.inverse_transform(prediction)[0])
```

```python
import numpy as np
from sklearn.naive_bayes import BernoulliNB
from sklearn.preprocessing import LabelEncoder

# Features (Has Coupon, Free Shipping)
X = np.array([
    [1, 1],
    [1, 0],
    [0, 1],
    [0, 0]
])

y_text = ["Purchase", "Purchase", "Ignore", "Ignore"]

le = LabelEncoder()
y = le.fit_transform(y_text)

model = BernoulliNB()
model.fit(X, y)

prediction = model.predict([[0, 1]])

print(le.inverse_transform(prediction))

# Predict for new dish: Has Coupon=1, Free Shipping=0
new_dish = [[1, 0]]
prediction = model.predict(new_dish)

# Decode prediction
print("Prediction:", le.inverse_transform(prediction)[0])
```
