```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

# Dataset
# Weight (g), Texture (0 = Smooth, 1 = Rough)
X = np.array([
    [150, 0],  # mango
    [170, 0],  # mango
    [140, 1],  # banana
    [130, 1]   # banana
])

# Labels
y = np.array(["mango", "mango", "banana", "banana"])

# Create KNN model (k = 3)
model = KNeighborsClassifier(n_neighbors=3)

# Train model
model.fit(X, y)

# Predict new fruit
new_fruit = [[160, 0]]   # Weight=160g, Texture=Smooth
prediction = model.predict(new_fruit)

print("Predicted Fruit:", prediction[0])
```
